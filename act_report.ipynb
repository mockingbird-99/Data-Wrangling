## RESULTS FROM DATA ANALYSIS AND VISUALIZATION- WeRateDogs
Dogs! Lovable creatures aren’t they! I analyzed the WeRateDogs twitter data set containing information on tweets about dogs, and the results are quite informative. 

Three insights are captured from the analysis:
1.	What are 3 the most popular dog names?
2.	Which dog classes are most common?
3.	What is the most common dog type in the p2 column?

The data set after cleaning contained 2073 records of data for 21 different columns of features. I analyzed the features in the above-mentioned bullets to arrive at the desired answers.
Each insight is expounded and visualized in the following sections:

1.	What are the 3 most popular dog names?
The `value_counts()` function was extremely helping in the analyses. To obtain the 3 most popular dog names in our data sets, we used the `value_counts()` function to order the dog names in descending order of frequency. The following were obtained
Charlie    11
Oliver     10
Tucker     10
Lucy       10
Cooper     10
Penny      10

Placing Charlie on top of the list, and **Oliver, Tucker, Lucy, Cooper** and **Penny** next. Below is a bar chart of the top 3 dog names, selecting Tucker and Oliver from the runners up.


 
2.	Which dog classes are most common?
For this analysis, we get our results from analyzing the `dogtionary` column on the data set. Applying `value_counts()` again provides us with the following order:
pupper     210
doggo       67
puppo       23
mixed       13
floofer      7

Placing **pupper** at first place, followed by **doggo, puppo, mixed** and then **floofer**. The mixed breeds fall in 2 breeds (e.g., doggopuppo)

 
3.	Which is the most common dog type in the p2 column?
The `p2` column contains the algorithm’s second most likely prediction. The descending order of frequency of this list is as follows:
labrador_retriever                104
golden_retriever                   92
cardigan                           74
chihuahua                          44
pomeranian                         42

These represent the top 5 most likely predictions, with the **labrador_retriever** atop the list with 104, and the **golden_retreiver** next with 92.
